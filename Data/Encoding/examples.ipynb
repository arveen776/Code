Example of encoding converting the bite --> numbers --> human ( text)

```python
b1 = 0b01000010  # 'B'
b2 = 0b00100000  # ' ' (space)
b3 = 0b01000001  # 'A'

data = bytes([b1, b2, b3])

print(data) 
print(data.decode())  
```

Now going in reverse human( text) --> numbers --> bites

```python
text = "What i look like"

encode = text.encode()

print(list(encode))

for byte in encode:
    print(f"{byte:08b}")
```
